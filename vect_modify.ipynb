```python
from matplotlib import pyplot as plt
import numpy as np

plt.figure(figsize=(5, 5))
plt.axis('equal')

# задаем точки квадрата
# поскольку будем рисовать как обычный график, то последняя точка [0,0] нужна, чтобы замкнуть наш квадрат 
orig_vect = np.array(
    [ 
        [0, 0], 
        [1, 0], 
        [1, 1], 
        [0, 1], 
        [0, 0]
    ])
print("Оригинальный векторы:\n", orig_vect)

plt.plot(orig_vect[:, 0], orig_vect[:, 1], label='исходный квадрат') # рисуем оригинальный квадрат. Передаем в plt.plot массив ординат и массив абсцисс

modify_matrix = np.array(
    [
        [2, 0], 
        [0, 2]
    ])
print("Матрица преобразования:\n", modify_matrix)

modified_vect = orig_vect.dot(modify_matrix)
print("Преобразованные векторы:\n",modified_vect)

plt.plot(modified_vect[:, 0], modified_vect[:, 1], label='преобразованный квадрат')
plt.legend()

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
from math import sin, cos, radians

plt.figure(figsize=(5, 5))
plt.axis('equal')

# задаем точки квадрата
# поскольку будем рисовать как обычный график, то последняя точка [0,0] нужна, чтобы замкнуть наш квадрат 
orig_vect = np.array(
    [ 
        [0, 0], 
        [1, 0], 
        [1, 1], 
        [0, 1], 
        [0, 0]
    ])
print("Оригинальный векторы:\n", orig_vect)

plt.plot(orig_vect[:, 0], orig_vect[:, 1], label='исходный квадрат') # рисуем оригинальный квадрат. Передаем в plt.plot массив ординат и массив абсцисс

rotate_angel = 45
rotate_radians = radians(rotate_angel)
modify_matrix = np.array(
    [
        [cos(rotate_radians), -sin(rotate_radians)], 
        [sin(rotate_radians), cos(rotate_radians)]
    ])
print("Матрица преобразования:\n", modify_matrix)

modified_vect = orig_vect.dot(modify_matrix)
print("Преобразованные векторы:\n",modified_vect)

plt.plot(modified_vect[:, 0], modified_vect[:, 1], label='преобразованный квадрат')
plt.legend()

plt.show()
```
